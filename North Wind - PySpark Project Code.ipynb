## Import Libraries

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('NorthWind').getOrCreate()
from pyspark.sql.functions import *
```

## Load the data

### Products Table

```
products = spark.read.csv("/FileStore/tables/Products.csv",inferSchema=True,header=True)
products.display()
```

### Orders Table

```
orders = spark.read.csv("/FileStore/tables/Orders.csv",inferSchema=True,header=True)
orders.display()
```

### Customers Table

```
customers = spark.read.csv("/FileStore/tables/Customers.csv",inferSchema=True,header=True)
customers.display()
```

### Employees Table

```
employees = spark.read.csv("/FileStore/tables/Employees.csv",inferSchema=True,header=True)
employees.display()
```

### Suppliers

```
suppliers = spark.read.csv("/FileStore/tables/Suppliers.csv",inferSchema=True,header=True)
suppliers.display()
```

## Questions

### 1. How many current products cost less than $20?

```
products.filter((col("UnitPrice")<20) & (col("Discontinued")==False)).agg(count('ProductID').alias('TotalCount')).display()
```

### 2. Which product is most expensive?

```
products.orderBy(col("UnitPrice").desc()).limit(1).show()
```

### 3. What is average unit price for our products

```
products.filter(col("UnitPrice").isNull()).count()
products.agg(round(avg("UnitPrice"),2)).display()
```

### 4. How many products are above the average unit price?

```
avgp=products.agg(round(avg("UnitPrice"),2).alias('Avg')).collect()[0]["Avg"]
products.filter((col("UnitPrice"))>avgp).display()
products.filter((col("UnitPrice"))>avgp).agg(count("ProductId")).display()
```

### 5. How many products cost between $15 and $25 (inclusive)?

```
products.filter((col('UnitPrice')>=15) & (col('UnitPrice')<=25)).display()
```


### 6. What is the average number of products (not qty) per order?

```
avgp=orders.groupBy('OrderID').agg(count("ProductID").alias('AvgProduct'))
avgp.agg(ceil(avg('AvgProduct'))).display()
```

### 7. What is Order Value in $ of open orders? (Not Shipped Yet)

```
orders.filter(col("ShippedDate").isNull()).agg(sum("_Sales")).display()
```

### 8. How many orders are 'single item' (only one product ordered)?

```
avgp=orders.groupBy('OrderID').agg(count("ProductID").alias("AVGPRODUCT"))
avgp.filter(col("AVGPRODUCT")==1).agg(count("OrderID")).display()
```

### 9. Average Sales per Transaction (orderID) for "Romero y Tomillo"

```
orders.filter(col("ShipName")=="Romero y tomillo").groupby('OrderID').agg(avg('_Sales')).display()
```

### 10. How many days since "North/South" last purchase?

```

cust=customers.filter(col("CompanyName")=="North/South").select("CustomerID").collect()[0]["CustomerID"]
df=orders.filter(col("CustomerID")==cust).orderBy(col("OrderDate").desc()).limit(1)
df=df.withColumn("CurrDate",current_date())
df.withColumn('Diff',datediff(col("CurrDate"),col("OrderDate"))).select("Diff").display()
```

### 11. How many customers have ordered only once?

```
c = customers.alias("c")
o = orders.alias("o")

df=c.join(o,c.CustomerID==o.CustomerID,"inner").distinct().groupBy("c.CustomerID").agg(countDistinct("OrderID").alias("Count"))
df.filter(col("Count")==1).display()
```

### 12. How many new customers in 2023?

```
from pyspark.sql.window import Window

order_df=orders.select("OrderId","CustomerId","OrderDate").distinct()
window = Window.partitionBy("CustomerId").orderBy("OrderDate")
order_df=order_df.withColumn('Ranking',rank().over(window))
order_df.filter((col('Ranking')==1) & (year(col('OrderDate'))==2023)).display()
```

### 13. How many Lost Customers in 2023?

```

df2023 = orders.filter(year(col('OrderDate')) == 2023).select('CustomerId').distinct()
df2024 = orders.filter(year(col('OrderDate')) == 2024).select('CustomerId').distinct()

lost_customers = df2023.subtract(df2024)

lost_customers.display()
```


### 16. How many products are out of stock?

```
df=products.filter(col('UnitsInStock')==0)
df.agg(countDistinct('ProductID')).display()
```

### 17. How many products need to be restocked? (based on restock levels)

```
products.filter(col('UnitsInStock')<=col('ReorderLevel')).display()
```

### 18. How many products on order we need to restock?

```
products.filter((col('UnitsInStock')<=col('UnitsOnOrder')) & (col('UnitsOnOrder')>0)).display()
```

### 19. What is the stocked value of the discontinued products?

```
products.filter(col('Discontinued')==True).select('ProductID','UnitsInStock').display()
```

### 20. Which vendor has the highest stock value?

```
products.join(suppliers,products.SupplierID==suppliers.SupplierID,"inner").distinct().withColumn('StockValue',col('UnitsInStock')*col('UnitPrice')).groupBy('CompanyName').agg(sum('StockValue')).orderBy(col('sum(StockValue)').desc()).limit(1).display()
```


### 21. How many employees are female?

```
employees.filter((col('Gender')=="Female") | (col('TitleOfCourtesy')=='Ms.')).display()
```


### 22. How many employees are 60 years old or over?

```

employees=employees.withColumn("BirthDate", to_date(col("BirthDate"), "dd-MM-yyyy"))
employees = employees.withColumn("Age", floor(datediff(current_date(), col("BirthDate")) / 365.25))
employees.select("EmployeeID", "BirthDate", "Age").filter(col('age')>=60).agg(count('EmployeeID')).display()
```

### 23. Which employee had the highest sales in 2022?

```

joined_df = orders.join(employees, on='EmployeeID', how='inner')

sales_by_employee = joined_df.filter(year('OrderDate')==2022).groupBy("EmployeeID") \
    .agg(sum("_Sales").alias("TotalSales")).orderBy(col('TotalSales').desc()).limit(1)
sales_by_employee.join(employees,on='EmployeeID',how='inner').display()
```

### 24. How many employees sold over $100k in 2023?

```
joined_df = orders.join(employees, on='EmployeeID', how='inner')

sales_by_employee = joined_df.filter(year('OrderDate')==2023).groupBy("EmployeeID") \
    .agg(sum("_Sales").alias("TotalSales")).orderBy(col('TotalSales')).filter(col('TotalSales')>100000).agg(count('EmployeeID')).display()
```

### 25. How many employees got hired in 1994?

```
employees=employees.withColumn("HireDate", to_date(col("HireDate"), "dd-MM-yyyy"))

employees.filter(year(col('HireDate'))==1994).agg(count('EmployeeID')).display()
```
